### Imports

```python
from common_language import _LANGUAGES
import processing as prlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Get datas

```python
data_f, train_df, test_df, validation_df = prlib.get_dataframes()
train_df['Length'] = train_df['paths'].apply(lambda x: prlib.get_length(prlib.get_path('train', x)))
test_df['Length'] = test_df['paths'].apply(lambda x: prlib.get_length(prlib.get_path('test', x)))
validation_df['Length'] = validation_df['paths'].apply(lambda x: prlib.get_length(prlib.get_path('validation', x)))
data_f = pd.concat([train_df, test_df, validation_df])
```

### Mean length

```python
new_data_f = data_f[['language', 'Length']].groupby('language').mean()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Mean Audio File Length by Language')
plt.xlabel('Languages')
plt.ylabel('Mean Length')
plt.xticks(rotation=90)

#plt.savefig('C:/Users/Samir/Documents/GitHub/Projet-STT-3795/images/mean_audio_length.svg', format='svg')  # Save as SVG
#plt.show()
#plt.close() 

plt.show()
```

### Number of audio by language

```python
new_data_f = data_f[['language', 'Length']].groupby('language').count()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Number of Audio by Language')
plt.xlabel('Languages')
plt.ylabel('Number of audio')
plt.xticks(rotation=90)

#plt.savefig('C:/Users/Samir/Documents/GitHub/Projet-STT-3795/images/num_of_audio.svg', format='svg')  # Save as SVG
#plt.show()
#plt.close()

plt.show()
```

### Sum of audio length by language

```python
new_data_f = data_f[['language', 'Length']].groupby('language').sum()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Sum Audio File Length by Language')
plt.xlabel('Languages')
plt.ylabel('Sum Length (seconds)')
plt.xticks(rotation=90)

#plt.savefig('C:/Users/Samir/Documents/GitHub/Projet-STT-3795/images/sum_of_audio.svg', format='svg')  # Save as SVG
#plt.show()
#plt.close() 

plt.show()
```
