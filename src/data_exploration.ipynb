### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow as pa
from mutagen.wave import WAVE
from datasets import load_dataset
```

### Get datas

```python
def get_length(path, type):
    audio = WAVE(f'./wav_files/{type}/' + path)
    audio_info = audio.info
    return audio_info.length

train_df = pd.read_csv('./wav_files/train/train_data.csv')
train_df['Length'] = train_df['paths'].apply(lambda x: get_length(x, 'train'))

test_df = pd.read_csv('./wav_files/test/test_data.csv')
test_df['Length'] = test_df['paths'].apply(lambda x: get_length(x, 'test'))

validation_df = pd.read_csv('./wav_files/validation/validation_data.csv')
validation_df['Length'] = validation_df['paths'].apply(lambda x: get_length(x, 'validation'))
```

```python
data_f = pd.concat([test_df, validation_df, train_df])
```

### Mean length

```python
new_train_df = train_df[['language', 'Length']].groupby('language').mean()
sns.barplot(x='language', y='Length', data=new_train_df, ci=None)
plt.title('Mean Audio File Length by Language for train data')
plt.xlabel('Languages')
plt.ylabel('Mean Length')
plt.xticks(rotation=90)
plt.show()
```

```python
new_data_f = data_f[['language', 'Length']].groupby('language').mean()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Mean Audio File Length by Language for test data')
plt.xlabel('Languages')
plt.ylabel('Mean Length')
plt.xticks(rotation=90)
plt.show()
```

### Number of audio by language

```python
new_train_df = train_df[['language', 'Length']].groupby('language').count()
sns.barplot(x='language', y='Length', data=new_train_df, ci=None)
plt.title('Number of audio by Language for train data')
plt.xlabel('Languages')
plt.ylabel('Number of audio')
plt.xticks(rotation=90)
plt.show()
```

```python
new_data_f = data_f[['language', 'Length']].groupby('language').count()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Number of Audio by Language for test data')
plt.xlabel('Languages')
plt.ylabel('Number of audio')
plt.xticks(rotation=90)
plt.show()
```

### Sum of audio length by language

```python
new_train_df = train_df[['language', 'Length']].groupby('language').sum()
sns.barplot(x='language', y='Length', data=new_train_df, ci=None)
plt.title('Sum Audio File Length by Language for train data')
plt.xlabel('Languages')
plt.ylabel('Sum Length (seconds)')
plt.xticks(rotation=90)
plt.show()
```

```python
new_validation_df = validation_df[['language', 'Length']].groupby('language').sum()
sns.barplot(x='language', y='Length', data=validation_df, ci=None)
plt.title('Sum Audio File Length by Language for validation data')
plt.xlabel('Languages')
plt.ylabel('Sum Length (seconds)')
plt.xticks(rotation=90)
plt.show()
```

```python
new_data_f = data_f[['language', 'Length']].groupby('language').sum()
sns.barplot(x='language', y='Length', data=new_data_f, ci=None)
plt.title('Sum Audio File Length by Language for test + validation')
plt.xlabel('Languages')
plt.ylabel('Sum Length (seconds)')
plt.xticks(rotation=90)
plt.show()
```
